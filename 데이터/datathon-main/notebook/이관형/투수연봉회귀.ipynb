```python
pitching.reset_index(drop=True, inplace=True)
```

```python
pitching
```

```python
trainSet = pitching[pitching['연도'] != 2024]
testSet = pitching[pitching['연도'] == 2024]
```

```python
x = trainSet.drop(columns=['Name','pid','연도','이후연봉'], axis=1)
y = trainSet['이후연봉']
x.columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = RandomForestRegressor(n_estimators=1000, random_state=42)
model.fit(X_train, y_train)

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))

preds = model.predict(X_test)

mse = mean_squared_error(y_test, preds)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, preds)
mape = np.mean(np.abs((y_test - preds) / y_test)) * 100

print(mse)
print(rmse)
print(mae)
print(mape)
```

```python
test_x = testSet.drop(columns= ['Name','pid','연도','이후연봉'])
test_y = testSet['이후연봉']

preds = model.predict(test_x)
result = testSet.copy()
result['예측연봉']=preds

# result = result[(result['연봉'] < 100001) & (result['이후연봉'] < 100000)]

result.to_csv('24투수_RF.csv', index=False)
```

```python
import pandas as pd



# 출력 행, 열, 너비, 컬럼명 길이 제한 해제
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

df7 = pd.read_parquet(r"C:\Users\rtc32\OneDrive\바탕 화면\final_project\train\3.승인매출정보\201807_train_승인매출정보.parquet")
df8 = pd.read_parquet(r"C:\Users\rtc32\OneDrive\바탕 화면\final_project\train\3.승인매출정보\201808_train_승인매출정보.parquet")
df9 = pd.read_parquet(r"C:\Users\rtc32\OneDrive\바탕 화면\final_project\train\3.승인매출정보\201809_train_승인매출정보.parquet")
df10 = pd.read_parquet(r"C:\Users\rtc32\OneDrive\바탕 화면\final_project\train\3.승인매출정보\201810_train_승인매출정보.parquet")
df11 = pd.read_parquet(r"C:\Users\rtc32\OneDrive\바탕 화면\final_project\train\3.승인매출정보\201811_train_승인매출정보.parquet")
df12 = pd.read_parquet(r"C:\Users\rtc32\OneDrive\바탕 화면\final_project\train\3.승인매출정보\201812_train_승인매출정보.parquet")
import os

#데이터 병합
dfs = [df7, df8, df9, df10, df11, df12]
merged_df = pd.concat(dfs, ignore_index=True)


save_dir = 'distplots'
os.makedirs(save_dir, exist_ok=True)

import matplotlib.pyplot as plt
import seaborn as sns
import gc
import random

num_cols = [col for col, dtype in merged_df.dtypes.items() if dtype in ['int64', 'float64']]

for col in num_cols:
    try:
        values = []
        sample_size = 2400000  
        count = 0
        for v in merged_df[col]:
            if pd.notnull(v):
                count += 1
                # reservoir sampling
                if len(values) < sample_size:
                    values.append(v)
                else:
                    r = random.randint(0, count - 1)
                    if r < sample_size:
                        values[r] = v
        if not values:
            print(f"⚠️ {col} 데이터 없음")
            continue
        data = pd.Series(values)
        if data.dtype == 'float64':
            data = data.astype('float32')
        elif data.dtype == 'int64':
            data = data.astype('int32')
        plt.figure(figsize=(8, 4))
        sns.histplot(data, bins=50, kde=False)
        plt.title(f'{col} 분포')
        plt.xlabel(col)
        plt.ylabel('빈도수')
        plt.tight_layout()
        plt.savefig(os.path.join(save_dir, f"{col}_histplot.png"))
        plt.close()
        gc.collect()
        print(f"✅ {col} 완료")
    except MemoryError:
        print(f"⚠️ MemoryError: {col} 컬럼 생략")
    except Exception as e:
        print(f"⚠️ {col} 에러 발생: {e}")
```

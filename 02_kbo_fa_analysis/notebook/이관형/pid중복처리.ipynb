수비기록
* G - 합계
* *GS - 합계
* IP - 합계(소수점이하 처리)
* TC - 합계
* PO - 합계
* Ass - 합계
* E - 합계
* F% - (PO+ASS) / (PO+ASS+E)
* RF9 - (TC-E) * 9 / IP
* RAA - 합
* /144 - 종합 RAA /144
* RAAwithPOS = 종합RAA + POSAdj
* WAAwoPOS = 종합RAA / RPW
* WAAwithPOS = RAAwithPOS / RPW

* 2024 RPW = 10.945

```python
import math

def sum_with_ternary_fraction(series):
    str_values = series.astype(str)
    splits = str_values.str.split('.', expand=True)
    int_values = splits[0].astype(int).sum()
    fraction_values = splits[1].astype(int).sum()
    fraction_values = fraction_values // 3

    return int_values + fraction_values
```

```python
RPW_2024 = 10.945
```

```python
이유찬_po = 이유찬_row['PO']
이유찬_ass = 이유찬_row['Ass']
이유찬_e = 이유찬_row['E']
이유찬_tc = 이유찬_row['TC']
이유찬_row['IP'] = sum_with_ternary_fraction(이유찬['IP'])
이유찬_row['F%'] = (이유찬_po + 이유찬_ass) / (이유찬_po + 이유찬_ass + 이유찬_e)
이유찬_row['RF9'] = (이유찬_tc - 이유찬_e) * 9 / 이유찬_row['IP']
이유찬_row['/144'] = 이유찬_row['종합 RAA'] * 144 / 이유찬_row['G']
이유찬_row['POSAdj'] = (이유찬['G'] * 이유찬['POSAdj']).sum() / 이유찬_row['G']
# 이유찬_row['POSAdj'] = 이유찬['POSAdj'].iloc[0] * 이유찬_row['G'] / 이유찬['G'].iloc[0]
이유찬_row['RAAwithPOS'] = 이유찬_row['종합 RAA'] + 이유찬_row['POSAdj']
이유찬_row['WAAwoPOS'] = 이유찬_row['종합 RAA'] / RPW_2024
이유찬_row['WAAwithPOS'] = 이유찬_row['RAAwithPOS'] / RPW_2024
이유찬_row['WAAwithPOS▼'] = 이유찬_row['RAAwithPOS'] / RPW_2024
```

```python
RPW_2024 = 10.945
RPW_2023 = 9.801
def unite_rows(df, pid):
    df_pid = df[df['ID']==pid].copy()
    df_pid.sort_values(by='G', ascending=False, inplace=True)
    print(df_pid['Name'].iloc[0])
    print(df_pid[['Team', '팀']])
    united_row = {
        'Rank': df_pid['Rank'].iloc[0],
        'Name': df_pid['Name'].iloc[0],
        'Team': df_pid['Team'].iloc[0],
        'G': df_pid['G'].sum(),
        'GS': df_pid['GS'].sum(),
        'TC': df_pid['TC'].sum(),
        'PO': df_pid['PO'].sum(),
        'Ass': df_pid['Ass'].sum(),
        'E': df_pid['E'].sum(),
        'Range RAA': df_pid['Range RAA'].sum(),
        'Arm RAA': df_pid['Arm RAA'].sum(),
        'Err RAA': df_pid['Err RAA'].sum(),
        'DP RAA': df_pid['DP RAA'].sum(),
        'Bt RAA': df_pid['CS RAA'].sum(),
        'Blk RAA': df_pid['Blk RAA'].sum(),
        'Frm RAA': df_pid['Frm RAA'].sum(),
        '종합 RAA': df_pid['종합 RAA'].sum(),
        '팀': df_pid['팀'].iloc[0],
        'pid': df_pid['pid'].iloc[0],
        'ID': df_pid['ID'].iloc[0]
    }
    pid_po = united_row['PO']
    pid_ass = united_row['Ass']
    pid_e = united_row['E']
    pid_tc = united_row['TC']
    united_row['IP'] = sum_with_ternary_fraction(df_pid['IP'])
    united_row['F%'] = (pid_po + pid_ass) / (pid_po + pid_ass + pid_e)
    united_row['RF9'] = (pid_tc - pid_e) * 9 / united_row['IP']
    united_row['/144'] = united_row['종합 RAA'] * 144 / united_row['G']
    united_row['POSAdj'] = (df_pid['G'] * df_pid['POSAdj']).sum() / united_row['G']
    # united_row['POSAdj'] = df_pid['POSAdj'].iloc[0] * united_row['G'] / df_pid['G'].iloc[0]
    united_row['RAAwithPOS'] = united_row['종합 RAA'] + united_row['POSAdj']
    united_row['WAAwoPOS'] = united_row['종합 RAA'] / RPW_2023
    united_row['WAAwithPOS'] = united_row['RAAwithPOS'] / RPW_2023
    united_row['WAAwithPOS▼'] = united_row['RAAwithPOS'] / RPW_2023

    for key, value in united_row.items():
        if type(value) == np.float64:
            united_row[key] = round(value, 3)

    df_row = pd.DataFrame(united_row, index=[0])
    df_tmp = df[df['ID'] != pid]
    df_tmp = pd.concat([df_tmp, df_row], ignore_index=True)
    print(f'united row success')
    return df_tmp
```

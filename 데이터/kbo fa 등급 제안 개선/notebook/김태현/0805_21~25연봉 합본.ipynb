```python
# 기본
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 경고 뜨지 않게 설정
import warnings
warnings.filterwarnings('ignore')

# 그래프 설정
sns.set()

# 그래프 기본 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False

# 데이터 전처리 알고리즘
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
# 1. 파일 불러오기
player_sal = pd.read_csv('선수 연봉 데이터(연봉).csv')
sal_data_21 = pd.read_csv('2021_연봉_statiz.csv')
sal_data_22 = pd.read_csv('2022_연봉_statiz.csv')
```

```python
# 연도별 데이터프레임 
sal_data_21['연도'] = 2021
sal_data_22['연도'] = 2022

display(player_sal.head())
display(sal_data_21.head())
display(sal_data_22.head())
```

```python
player_sal['연봉(만원)'].info()
```

```python
player_sal['연봉(만원)'] = player_sal['연봉(만원)'].astype(int)
```

```python
display(player_sal.head())
```

```python
# 파일 병합하기 
sal_total = pd.concat([sal_data_21, sal_data_22, player_sal], ignore_index=True)
sal_total
```

```python
df = sal_total.copy()
```

```python
df['연봉(만원)'].info()
```

```python
# 1. 일단 숫자(실수)로 변환
df['연봉(만원)'] = pd.to_numeric(df['연봉(만원)'], errors='coerce')

# 2. 결측값(NaN)을 0이나 원하는 값으로 처리 (필수 아님, 필요하면)
df['연봉(만원)'] = df['연봉(만원)'].fillna(0)

# 3. 정수(int)형으로 변환
df['연봉(만원)'] = df['연봉(만원)'].astype(int)
```

```python
# 1. 연봉 컬럼 복사 (이미 되어있으면 생략)
df['연봉'] = df['연봉(만원)']

# 2. 팀별 연도별 연봉순위 (1등 = 가장 연봉 높은 선수)
df['팀연봉순위'] = df.groupby(['연도', '팀'])['연봉'].rank(ascending=False, method='first')

# 3. 전체 연도별 연봉순위
df['전체연봉순위'] = df.groupby('연도')['연봉'].rank(ascending=False, method='first')

# 4. 연봉이 NaN인 선수 삭제
df = df.dropna(subset=['연봉'])
df
```

```python
df['팀연봉순위'] = df['팀연봉순위'].astype(int)
df['전체연봉순위'] = df['전체연봉순위'].astype(int)
```

```python
df.drop(columns='연봉', inplace=True)
df
```

```python
df.to_csv('2021_2025_연봉_합본데이터.csv', index=False)
```

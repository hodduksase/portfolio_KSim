```python
bat_21_data.rename(columns={'Sort▼': 'WAR', 'Team': '포지션'}, inplace=True)
bat_22_data.rename(columns={'Sort▼': 'WAR', 'Team': '포지션'}, inplace=True)

player_data_2124.rename(columns={'Sort▼': 'WAR'}, inplace=True)
```

```python
bat_21_data.drop(columns='포지션', inplace=True)
bat_22_data.drop(columns='포지션', inplace=True)

cols = ['선수', '팀'] + [col for col in bat_21_data.columns if col not in ['선수', '팀']]
bat_21_data = bat_21_data[cols]

cols_2 = ['선수', '팀'] + [col for col in bat_22_data.columns if col not in ['선수', '팀']]
bat_22_data = bat_22_data[cols_2]
```

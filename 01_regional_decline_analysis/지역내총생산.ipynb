```python
file_path = 'data/day3/시도별_경제활동별_지역내총생산_20250530102718.csv'
```

```python
# 파일 처음 두 줄을 따로 읽는다
with open(file_path, encoding='cp949') as f:
    header1 = f.readline().strip().split(",")
    header2 = f.readline().strip().split(",")

columns = []
for h1, h2 in zip(header1, header2):
    h1 = h1.strip()
    h2 = h2.strip()
    if h1 == h2:
        columns.append(h1)
    else:
        columns.append(f"{h1}_{h2}")

# 데이터 본문 읽기 (헤더 2줄 스킵)
df = pd.read_csv(file_path, skiprows=2, names=columns, encoding='cp949')
```

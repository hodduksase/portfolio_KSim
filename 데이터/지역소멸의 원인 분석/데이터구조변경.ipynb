```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv('data\경찰철_범죄발생지역별통계_2023.csv', encoding='cp949')

# melt
df_melted = pd.melt(df,
                    id_vars=['범죄대분류', '범죄중분류'],
                    var_name='지역',
                    value_name='발생건수')

# 시도/구군 분리 함수 정의
def split_region(region):
    if region.startswith('서울'):
        return '서울특별시', region.replace('서울', '')
    elif region.startswith('부산'):
        return '부산광역시', region.replace('부산', '')
    elif region.startswith('대구'):
        return '대구광역시', region.replace('대구', '')
    elif region.startswith('인천'):
        return '인천광역시', region.replace('인천', '')
    elif region.startswith('광주'):
        return '광주광역시', region.replace('광주', '')
    elif region.startswith('대전'):
        return '대전광역시', region.replace('대전', '')
    elif region.startswith('울산'):
        return '울산광역시', region.replace('울산', '')
    elif region.startswith('세종'):
        return '세종특별자치시', region.replace('세종', '')
    elif region.startswith('경기'):
        return '경기도', region.replace('경기', '')
    elif region.startswith('강원'):
        return '강원도', region.replace('강원', '')
    elif region.startswith('충북'):
        return '충청북도', region.replace('충북', '')
    elif region.startswith('충남'):
        return '충청남도', region.replace('충남', '')
    elif region.startswith('전북'):
        return '전라북도', region.replace('전북', '')
    elif region.startswith('전남'):
        return '전라남도', region.replace('전남', '')
    elif region.startswith('경북'):
        return '경상북도', region.replace('경북', '')
    elif region.startswith('경남'):
        return '경상남도', region.replace('경남', '')
    elif region.startswith('제주'):
        return '제주특별자치도', region.replace('제주', '')
    else:
        return '기타', region  # 알 수 없는 경우 처리

# 시도/구군 컬럼 생성
df_melted[['시도', '구군']] = df_melted['지역'].apply(lambda x: pd.Series(split_region(x)))

# 저장
df_melted.to_csv('태블로용_데이터_시도구군분리.csv', index=False)
```

```python
df_melted['시군구'] = df_melted['시도'] + ' ' + df_melted['구군']
df_melted.to_csv('태블로용_데이터_시군구.csv', index=False)
```

```python
def fix_gyeonggi(region):
    if region.startswith('경기도 도'):
        return '경기도 ' + region[len('경기도 도'):]
    elif region.startswith('경기도'):
        return region
    elif region.startswith('경기'):
        return '경기도 ' + region[len('경기'):]
    else:
        return region

df_melted['시군구'] = df_melted['시군구'].apply(fix_gyeonggi)
```

```python
df = pd.read_csv('시군구별 인구이동자수, 연도기준 (2013 - 2024) .csv', encoding='cp949')

# 공백 제거
df['행정구역(시군구)별'] = df['행정구역(시군구)별'].str.strip()

# 시도 목록 (전국 시도 기준)
sido_names = [
    '서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시',
    '대전광역시', '울산광역시', '세종특별자치시', '경기도', '강원특별자치도',
    '충청북도', '충청남도', '전북특별자치도', '전라남도', '경상북도', '경상남도',
    '제주특별자치도'
]

# 시도/시군구 저장용 리스트
sido_list = []
sigungu_list = []

current_sido = None

for name in df['행정구역(시군구)별']:
    if name in sido_names:
        current_sido = name
        sido_list.append(current_sido)
        sigungu_list.append(None)
    else:
        sido_list.append(current_sido)
        sigungu_list.append(name)

# 컬럼 추가
df['시도'] = sido_list
df['시군구'] = sigungu_list

# 시군구가 있는 행만 추출
df_result = df[df['시군구'].notna()].reset_index(drop=True)
```

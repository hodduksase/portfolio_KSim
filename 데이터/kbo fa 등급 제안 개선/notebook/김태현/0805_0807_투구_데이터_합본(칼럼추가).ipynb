### 파생변수 생성(선발, 중간, 마무리) 

```python
# 연도별로 분리 
df_pit_all_21 = df_pit_all[df_pit_all['연도'] == 2021]
df_pit_all_22 = df_pit_all[df_pit_all['연도'] == 2022]
df_pit_all_23 = df_pit_all[df_pit_all['연도'] == 2023]
df_pit_all_24 = df_pit_all[df_pit_all['연도'] == 2024]
```

```python
# 팀 별로 분리 
team_list_21 = df_pit_all_21['팀'].unique()
team_df_dict_21 = {team: df_pit_all_21[df_pit_all_21['팀'] == team] for team in team_list_21}
```

```python
# 1. 팀 이름 리스트
team_names = [
    'KIA 타이거즈', '두산 베어스', '롯데 자이언츠', '삼성 라이온즈', 'SSG 랜더스',
    'NC 다이노스', 'LG 트윈스', '키움 히어로즈', 'KT 위즈', '한화 이글스'
]
```

```python
# 2. 연도별 데이터프레임 리스트, 딕셔너리 리스트
year_dfs = [df_pit_all_21, df_pit_all_22, df_pit_all_23, df_pit_all_24]
team_dicts = []

for df in year_dfs:
    team_dict = {team: df[df['팀'] == team].copy() for team in team_names}
    team_dicts.append(team_dict)
```

```python
# 3. 투수 포지션 라벨링 함수
def label_pitchers(team_df):
    team_df['GS'] = pd.to_numeric(team_df['GS'], errors='coerce').fillna(0).astype(int)
    team_df['S'] = pd.to_numeric(team_df['S'], errors='coerce').fillna(0).astype(int)
    
    # 선발: GS 10이상, 최대 6명
    starters = (
        team_df[team_df['GS'] >= 10]
        .sort_values('GS', ascending=False)
        .head(6)
        .index
    )
    
    # 마무리: SV 10이상, 최대 3명
    closers = (
        team_df[team_df['S'] >= 10]
        .sort_values('S', ascending=False)
        .head(3)
        .index
    )
    
    # 기본: 중간계투
    team_df['투수세부분류'] = '중간계투'
    team_df.loc[starters, '투수세부분류'] = '선발'
    team_df.loc[closers, '투수세부분류'] = '마무리'
    return team_df

# 4. 실제 적용 및 저장
for i, team_dict in enumerate(team_dicts):
    for team in team_names:
        team_df = team_dict[team]
        team_dict[team] = label_pitchers(team_df)
```

```python
all_df_list = []
years = [2021, 2022, 2023, 2024]
```

```python
for i, team_dict in enumerate(team_dicts):
    for team in team_names:
        team_df = team_dict[team].copy()
        team_df['연도'] = years[i] 
        all_df_list.append(team_df)

# 하나로 병합하기 
df_all_pitchers = pd.concat(all_df_list, ignore_index=True)
```

```python
df_all_pitchers
```

```python
# 원핫 인코딩 직접
df_all_pitchers['선발'] = (df_all_pitchers['투수세부분류'] == '선발').astype(int)
df_all_pitchers['중간계투'] = (df_all_pitchers['투수세부분류'] == '중간계투').astype(int)
df_all_pitchers['마무리'] = (df_all_pitchers['투수세부분류'] == '마무리').astype(int)

# 기존 칼럼 삭제
df_all_pitchers = df_all_pitchers.drop(columns=['투수세부분류'])
```

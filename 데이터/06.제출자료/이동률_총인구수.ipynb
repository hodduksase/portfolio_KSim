```python
import pandas as pd
import numpy as np

# 1. 데이터 불러오기
file_path = 'data/인구이동_연도데이터.csv'
df = pd.read_csv(file_path, encoding='utf-8-sig')
```

```python
# 원본 데이터를 담은 데이터 프레임(df)
df
```

원본 데이터를 복제하여 위와 같은 구조로 변경한다.

```python
# 6. '전국' 데이터 제외
# '전국'이 포함된 행을 제거한 새 데이터프레임 만들기
filtered_df = df[(df['시도'] != '전국') & (df['시군구'] != '전국')]
```

```python
print(filtered_df.head())
```

```python
# 이름이 길어서 다시 df에 필터링한 데이터프레임을 담겠습니다 
# drop=True는 전국을 삭제하면서 인덱스가 8번부터 시작하는걸 다시 0으로 땡겨옴
df = filtered_df.reset_index(drop=True)
```

```python
print(df.head())
```

```python
# 7. 항목 분류 함수 정의 및 적용

# a. 연도 컬럼 이름 정리: '2013 년' → '2013'
df.columns = [col.strip().replace(' 년', '') if '년' in col else col for col in df.columns]
# df
```

```python
# 단위 컬럼 삭제
df = df.drop(columns='단위')
# df
```

```python
# b. melt로 '연도' 컬럼 생성
df_melted = df.melt(id_vars=['시도', '시군구', '항목'], var_name='연도', value_name='값')
df_melted
```

```python
print(df_melted['항목'].dtype)
```

```python
import re
# [명] 제거하기
# 항목 컬럼 내 총전입[명] -> 총전입 으로 이름 변경
df_melted['항목'] = df_melted['항목'].str.replace(r'\[[^\]]*\]', '', regex=True).str.strip()
```

```python
df_melted
```

```python
# c. 필요한 항목만 필터링 (총전입, 총전출, 순이동)
df_filtered = df_melted[df_melted['항목'].isin(['총전입', '총전출', '순이동'])]
df_filtered
```

```python
df_filtered
```

```python
print(df_filtered['연도'].dtype)
print(df_filtered['연도'].unique())

df_filtered['연도'] = df_filtered['연도'].astype(int)
```

```python
print(df_filtered['연도'].dtype)
```

```python
print(df_filtered.head(10))
```

```python
df_pivot = df_filtered.pivot_table(
    index=['시도', '시군구', '연도'],
    columns='항목',
    values='값',
    aggfunc='sum',
    fill_value=0
).reset_index()

print(df_pivot['시도'].unique())
```

```python
# 원하는 순서로 시도 리스트 만들기
시도_순서 = [
    '서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시',
    '대전광역시', '울산광역시', '세종특별자치시',
    '경기도', '강원특별자치도', '충청북도', '충청남도',
    '전북특별자치도', '전라남도', '경상북도', '경상남도', '제주특별자치도'
]

# 카테고리형으로 바꿔서 정렬 기준 지정
df_pivot['시도'] = pd.Categorical(df_pivot['시도'], categories=시도_순서, ordered=True)

# 정렬
df_pivot = df_pivot.sort_values(['시도', '시군구', '연도']).reset_index(drop=True)
```

```python
# 1. '값'이 0.0인 행 제거
df_filtered = df_filtered[df_filtered['값'] != 0.0]

# 2. 이후 피벗
df_pivot = df_filtered.pivot_table(
    index=['시도', '시군구', '연도'],
    columns='항목',
    values='값'
).reset_index()

# 3. 컬럼명 정리 (총전입 → 전입 등)
df_pivot = df_pivot.rename(columns={'총전입': '전입', '총전출': '전출', '순이동': '순이동'})
```

```python
df_pivot['시군구'] = df_pivot['시군구'].str.replace(' ', '', regex=False)
```

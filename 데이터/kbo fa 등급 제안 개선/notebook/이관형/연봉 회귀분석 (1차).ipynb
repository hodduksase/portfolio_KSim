```python
x = df.drop(columns=['Name_x', 'pid', '연봉(만원)'])
y = df['연봉(만원)']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
x['pos'] = le.fit_transform(x['포지션'])
x['team'] = le.fit_transform(x['팀'])
```

```python
x.drop(columns=['포지션', '팀'], inplace=True)
x.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit_transform(x)
```

```python
import numpy as np

logy = np.log1p(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, logy, test_size=0.2, random_state=42)
```

```python
# Lasso
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.01)
lasso.fit(X_train, y_train)
```

```python
# ElasticNet
from sklearn.linear_model import ElasticNet

elastic = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic.fit(X_train, y_train)
```

```python
# SVR
from sklearn.svm import SVR

svr = SVR(kernel='linear', C=1.0, epsilon=0.2)
svr.fit(X_train, y_train)
```

```python
svr_poly = SVR(kernel='poly', C=1.0, epsilon=0.2)
svr_poly.fit(X_train, y_train)
```

```python
svr_rbf = SVR(kernel='rbf', C=0.01, epsilon=0.2, gamma=100)
svr_rbf.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

lasso_pred = lasso.predict(X_test)
mse = mean_squared_error(y_test, lasso_pred)
print(mse)
score = lasso.score(X_test, y_test)
print(score)
```

```python
elastic_pred = elastic.predict(X_test)
mse = mean_squared_error(y_test, elastic_pred)
print(mse)
score = elastic.score(X_test, y_test)
print(score)
```

```python
svr_pred = svr.predict(X_test)
mse = mean_squared_error(y_test, svr_pred)
print(mse)
score = svr.score(X_test, y_test)
print(score)
```

```python
poly_pred = svr_poly.predict(X_test)
mse = mean_squared_error(y_test, poly_pred)
print(mse)
score = svr_poly.score(X_test, y_test)
print(score)
```

```python
rbf_pred = svr_rbf.predict(X_test)
mse = mean_squared_error(y_test, rbf_pred)
print(mse)
score = svr_rbf.score(X_test, y_test)
print(score)
```

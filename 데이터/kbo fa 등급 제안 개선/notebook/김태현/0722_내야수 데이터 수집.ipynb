```python
import pandas as pd
import requests
```

```python
url = "https://statiz.sporki.com/stats/?m=main&m2=all&m3=default&so=WAR&ob=DESC&year=2024&po=13&lt=10100"
headers = {"User-Agent":"Mozilla/5.0"}
```

```python
response = requests.get(url, headers=headers)
response.encoding = "utf-8"   # 한글 사이트에서 꼭 필요!
```

```python
# 0번 테이블이 선수 스탯 (실제로 출력해보고 번호 확인!)
tables = pd.read_html(response.text)
df = tables[0]

# 컬럼명/데이터 살짝 확인
print(df.head())
```

```python
def convert_to_ymd_str_or_keep_etc(val):
    if pd.isna(val):
        return pd.NA
    elif val == '기타':
        return '기타'
    else:
        try:
            ymd_str = str(int(float(val)))  # '20211121'
            year = ymd_str[:4]
            month = ymd_str[4:6]
            day = ymd_str[6:8]
            return f"{year}-{month}-{day}"
        except:
            return pd.NA


df_filtered['RV신청일자'] = df_filtered['RV신청일자'].apply(convert_to_ymd_str_or_keep_etc)
```

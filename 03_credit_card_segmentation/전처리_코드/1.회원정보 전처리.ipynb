```python
def convert_to_year_month_str_or_keep_etc(val):
    if pd.isna(val):
        return pd.NA
    elif val == '기타':
        return '기타'
    else:
        try:
            ym_str = str(int(float(val))) 
            year = ym_str[:4]
            month = ym_str[4:]
            return f"{year}-{month}"
        except:
            return pd.NA


df1_filtered['최종유효년월_신용_이용가능'] = df1_filtered['최종유효년월_신용_이용가능'].apply(convert_to_year_month_str_or_keep_etc)
df1_filtered['최종유효년월_신용_이용'] = df1_filtered['최종유효년월_신용_이용'].apply(convert_to_year_month_str_or_keep_etc)
```

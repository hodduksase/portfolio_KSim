```python
# 날짜(YYYYMMDD) 또는 '사용하지 않음'(예: 10101) 패턴을 가진 int 컬럼 자동 탐지 및 전처리

def is_date_or_unused_col(series):
    # int 타입, 값이 100000 이하(사용안함) 또는 8자리(YYYYMMDD)인 값이 많으면 True
    if not np.issubdtype(series.dtype, np.integer):
        return False
    vals = series.dropna().unique()
    # 8자리(YYYYMMDD)와 100000 이하 값이 전체의 90% 이상이면 해당 패턴으로 간주
    date_like = sum((v > 10000000) & (v < 30000000) for v in vals)
    unused_like = sum(v == 10101 for v in vals)
    ratio = (date_like + unused_like) / len(vals) if len(vals) > 0 else 0
    return ratio > 0.9

for col in df.columns:
    if is_date_or_unused_col(df[col]):
        def parse_date_or_unused(x):
            if x == 10101:
                return pd.Timestamp('2262-04-11')
            try:
                return pd.to_datetime(str(int(x)), format='%Y%m%d')
            except:
                return np.nan
        df[col] = df[col].apply(parse_date_or_unused)
        print(f"{col} 컬럼을 날짜/2262-04-11 패턴으로 자동 전처리")
```

```python
import pandas as pd
import numpy as np
```

```python

batter = pd.read_csv('24타자_RF (1).csv')
pitcher = pd.read_csv('24투수_RF.csv')
```

```python
batter.columns
```

```python
pitcher.columns
```

```python
player = batter[['선수','연도','기존pid','연봉','이후연봉','예측연봉']]
player['포지션'] = np.argmax(batter[['포수','1루수','2루수','3루수','유격수','좌익수','중견수','우익수']].values, axis=1)
```

```python
player['포지션'] = player['포지션'].map({0:'포수',1:'1루수',2:'2루수',3:'3루수',4:'유격수',5:'좌익수',6:'중견수',7:'우익수'})
player['포지션']
```

```python
player
```

```python
pitcherA = pitcher[['Name','연도','pid','연봉','이후연봉', '예측연봉']]
pitcherA['포지션'] = '투수'
```

```python
pitcherA.rename(columns={'Name': '선수','pid':'기존pid'}, inplace=True)
```

```python
player = pd.concat([player, pitcherA])
player
```

```python
obatter = pd.read_csv('타자종합.csv')
opitcher = pd.read_csv('투수종합.csv')
```

```python
obatter = obatter[obatter['연도'] == 2024]
opitcher = opitcher[opitcher['연도'] == 2024]
obatter
```

```python
salary = pd.read_csv('pid_연봉 (1).csv')
salary
```

```python
salary_melt = salary.melt(id_vars='pid', var_name='연도', value_name='연봉')
salary_melt
```

```python
salary_melt
```

```python
salary_melt.rename(columns={'pid': '기존pid'}, inplace=True)
salary_melt
```

```python
salary_melt['연도'] = salary_melt['연도'].astype(int)
obatter = pd.merge(obatter, salary_melt, on=['기존pid', '연도'], how='left')
obatter
```

```python
obatter = obatter.iloc[:,:-2]
obatter
```

```python
obatter.rename(columns={'연봉_x': '연봉'}, inplace=True)
obatter
```

```python
salary_melt['연도'] = salary_melt['연도'] - 1
salary_melt.rename(columns={'연봉': '이후연봉'}, inplace=True)
obatter = pd.merge(obatter, salary_melt, on=['기존pid', '연도'], how='left')
obatter
```

```python
obatter.dropna(subset=['이후연봉'], inplace=True)
obatter
```

```python
opitcher.dropna(subset=['이후연봉'], inplace=True)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

def calculate_vif(X):
    X = add_constant(X)  # 상수항 추가 (intercept)
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data
```

```python
sns.kdeplot(y_test, label='실제')
sns.kdeplot(xgb_preds, label='xgb')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=xgb_preds, alpha=0.4)
plt.plot([y_test.min(), y_test.max()], [xgb_preds.min(), xgb_preds.max()], 'r--', label='Ideal')
plt.xlabel("실제 연봉")
plt.ylabel("예측 연봉")
plt.title("실제 vs 예측 산점도")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
residuals = y_test - xgb_preds

plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True, bins=50, color='purple')
plt.axvline(0, color='red', linestyle='--')
plt.title("잔차 분포")
plt.xlabel("잔차 (실제 - 예측)")
plt.ylabel("빈도")
plt.tight_layout()
plt.show()
```

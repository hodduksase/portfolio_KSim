```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

def calculate_vif(X):
    X = add_constant(X)  # 상수항 추가 (intercept)
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data
```

```python
param_train, param_test, target_train, target_test = train_test_split(x_train, y_train, test_size=0.15, random_state=42)
```

```python
import optuna

def objective(trial):
    params = {
        'n_estimators': 1000,
        'objective': 'regression',
        'metric': 'rmse',
        'verbosity': -1,
        'bagging_freq': 1,
        'random_state': 42,
        'learning_rate': trial.suggest_float('learning_rate', 1e-5, 1e-3, log=True),
        'num_leaves': trial.suggest_int('num_leaves', 2, 2**5),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'min_data_in_leaf': trial.suggest_int('min_data_in_leaf', 1, 100)
    }

    model = lgb.LGBMRegressor(**params)
    model.fit(param_train, target_train)
    y_pred = model.predict(param_test)
    rmse = mean_squared_error(target_test, y_pred)
    return np.sqrt(rmse)

study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=100)
```

```python
e24 = pd.read_csv('2024_연봉_statiz.csv')
```

```python
team_code = {
 'KIA 타이거즈': 0,
 '두산 베어스': 1,
 '롯데 자이언츠': 2,
 '삼성 라이온즈': 3,
 'SSG 랜더스': 4,
 'NC 다이노스': 5,
 'LG 트윈스': 6,
 '키움 히어로즈': 7,
 'KT 위즈': 8,
 '한화 이글스': 9,
}
```

```python
e24['tid'] = e24['팀'].map(team_code)
```

```python
e24
```

```python
e24['ID'] =  e24['pid'] * 10 + e24['tid']
e24['ID'].value_counts()
```

```python
e24.drop(columns=['pid', 'tid', '팀'], inplace=True)
```

```python
e24['pid'] = e24['ID']
e24.drop(columns=['ID'], inplace=True)
e24
```

```python
b24.drop(columns=['Rank', 'Name', 'Team'], inplace=True)
e24 = e24[['연봉(만원)','pid']]
e24
```

```python
t24 = pd.merge(f24, b24, on='pid')
t24 = pd.merge(t24, e24, on='pid')
t24
```

```python
t24.drop(columns='Rank', inplace=True)
```
